```python
def clean(text):
    try:
        import nltk
        import re
        nltk.download('stopwords')
        from nltk.corpus import stopwords
        import string
        stopword=set(stopwords.words('english'))
        text = text.lower()
        text = re.sub('\[.*?\]', '', text)
        text = re.sub('https?://\S+|www\.\S+', '', text)
        text = re.sub('<.*?>+', '', text)
        text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
        text = re.sub('\n', '', text)
        text = re.sub('\w*\d\w*', '', text)
        result = [word + " " for word in text.split() if word not in stopword]
        result = "".join(result)
        return result
    except Exception:
        return Exception
```

# Generating Word Cloud

The frequency or importance of each word in a word cloud, a data visualisation technique for expressing text data, shows its frequency or relevance. Using a word cloud, significant textual data points can be highlighted. Word clouds are widely used for analyzing data from social network websites.

```python
data["Review"] = data["Review"].apply(clean)
```

```python
data['Review']
```

```python
text = " ".join(i for i in data.Review)
text
```

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(text)
 
# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show() 
```

# we are using VADER to text processing

Just Check this video link for better under stand using VADER - https://www.youtube.com/watch?v=Alu_cCXNS-k

VADER Stand For - VADER (Valence Aware Dictionary and sEntiment Reasoner) is a lexicon and rule-based sentiment analysis tool that is specifically attuned to sentiments expressed in social media. VADER uses a combination of A sentiment lexicon is a list of lexical features (e.g., words) which are generally labeled according to their semantic orientation as either positive or negative. VADER not only tells about the Positivity and Negativity score but also tells us about how positive or negative a sentiment is.

```python
nltk.download('vader_lexicon')
#we are declaring Sentiment Analyzer object it returns four output it takes text as a input based on the text 
#it returns pos -> postive neg -> negative neu -> neutral & compund 
sentiments = SentimentIntensityAnalyzer()
data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data["Review"]]
data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data["Review"]]
data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data["Review"]]
data = data[["Review", "Positive", "Negative", "Neutral"]]
print(data.head())
```

```python
x = sum(data["Positive"])
y = sum(data["Negative"])
z = sum(data["Neutral"])

def sentiment_score(a, b, c):
    if (a>b) and (a>c):
        print("Positive üòä ")
    elif (b>a) and (b>c):
        print("Negative üò† ")
    else:
        print("Neutral üôÇ ")
sentiment_score(x, y, z)
```
